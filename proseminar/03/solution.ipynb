## Run heat_stencil_1D/2D/3D them with multiple problem and machine sizes

`````python
import csv
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

labels = [["1D_seq", "1D_1p1", "1D_4p8", "1D_8p16"],
          ["2D_seq", "2D_1p1", "2D_4p8", "2D_8p16"],
          ["3D_seq", "3D_1p1", "3D_4p8", "3D_8p16"]]

filepath = 'data.csv'
`````

`````python
def getDataFor(labels):
    with open(filepath) as csv_file:
        data = []
        for label in labels:
            csv_reader = csv.reader(csv_file, delimiter=',')
            data.append([[int(row[1]),float(row[2])] for row in csv_reader if row[0] == label])
            csv_file.seek(0)
    return data
`````

`````python
plt.rcParams["figure.figsize"] = (30,8)


def plot(title, ax, labels, data):
    sizes = [i[0] for i in data[0]]

    x = np.arange(len(sizes))
    width = 0.15
    
    rects1 = ax.bar(x - (width+width/2), [i[1] for i in data[0]], width, label=labels[0])
    rects2 = ax.bar(x - width/2, [i[1] for i in data[1]], width, label=labels[1])
    rects3 = ax.bar(x + width/2, [i[1] for i in data[2]], width, label=labels[2])
    rects4 = ax.bar(x + width + width/2, [i[1] for i in data[3]], width, label=labels[3])

    ax.set_title(title)
    ax.set_ylabel('Execution time in s')
    ax.set_xticks(x)
    ax.set_xticklabels(sizes)
    ax.set_xlabel('Problem Size')
    ax.legend()

    def autolabel(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),  # 3 points vertical offset
                        textcoords="offset points",
                        ha='center', va='bottom')
`````

`````python
fig, ax = plt.subplots(1,3)

for i in range(0,len(labels)): plot("Heat stencil {}D".format(i+1), ax[i], labels[i], getDataFor(labels[i]))

fig.tight_layout()
plt.show()
`````

# Exercise 2

### Measure the speedup and efficiency of all three stencil codes for varying problem and machine sizes/mappings. 

### Speedup
- **absolut:** reference ts is the fastest sequential version
- **relativ:** reference ts is the fastest parallel version run sequentially
$$ speedup_p = \frac{t_s}{t_p} $$


### Efficency
$$ efficency = \frac{speedup_p}{p} $$

### Strong scalability, weak scalability
- **strong scalablity**: how the program scales with a fixed problem size
- **weak scalability**: how the program scales when keeping the problem size proportional

For our tests we used strong scalabilty since for weak scalabilty its not quiete clear which parameters should be scaled. We could choose to scale N (grid size) but this parameter scales not linear with computation size (1D, 2D, 3D) and thus it will be harder to compare results.

### Figure

`````python
def plot2(ax1, dimension):
    cores = [1,8,16]
    ax2 = ax1.twinx()
    for problemSize in range(0,3):
        label = labels[dimension]
        data = getDataFor(label)
        seq = data[0][problemSize][1]
        speedup = [seq / data[i][problemSize][1] for i in range(1, len(data))]
        problemSizes = [data[i][problemSize][0] for i in range(1,len(data))]
        efficency = [speedup[i]/cores[i] for i in range(0,3)]
        
        ax1.set_xlabel('number of processors')
        ax1.set_ylabel('Speedup')
        ax1.set_title("Heat stencil {}D".format(dimension+1))
        ax1.plot(cores, speedup, label="Speedup {}".format(problemSizes[problemSize]) , marker='o',  linewidth=3, color=['red','green', 'blue'][problemSize])
        
        ax2.set_ylabel("Efficency")
        ax2.plot( cores, efficency, label="Efficency {}".format(problemSizes[problemSize]), marker='o', color=['orange','aquamarine','chartreuse'][problemSize], linewidth=3)
        ax2.legend()
        ax1.legend()

fig, ax1 = plt.subplots(1,3)
for i in range(0,3): plot2(ax1[i], i)
fig.tight_layout() 
plt.rcParams["figure.figsize"] = (24,8)
plt.show() 

`````
